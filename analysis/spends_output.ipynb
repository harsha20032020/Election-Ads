```python
from collections import Counter
def count_frequencies(list_of_lists):
    flattened_list = [item for sublist in list_of_lists for item in set(sublist)]
    frequency_counter = Counter(flattened_list)
    return frequency_counter.most_common()

# Combine both columns into a single list of lists
combined_list_of_lists = df['spacy_ner'].tolist() + df['twit_ner'].tolist()
c1 = count_frequencies(combined_list_of_lists)
c1[:50]
```

```python
df['ad_delivery_start_time'] = pd.to_datetime(df['ad_delivery_start_time'], errors='coerce')
df['ad_delivery_stop_time'] = pd.to_datetime(df['ad_delivery_stop_time'], errors='coerce')
```

```python
def filter_ads_by_date_range(df,d1,d2):
    valid_indices = set()
    for single_date in pd.date_range(start=d1, end=d2):
        rows_on_date = df[(df['ad_delivery_start_time'] <= single_date) & (df['ad_delivery_stop_time'] >= single_date)]
        valid_indices.update(rows_on_date.index)
    filtered_df = df.loc[valid_indices].reset_index(drop=True)
    return filtered_df
```

```python
import re
import warnings
warnings.filterwarnings('ignore')
def extract_state(text):
    match = re.search(r'\(([^)]*)\)', text)
    if match:
        return match.group(1)
    else:
        return None
def process_data(df, text_column1, text_column2, int_column1, int_column2, search_string):
    search_words = search_string.lower().split()

    # Create a mask based on the condition that all search words are present
    mask = df.apply(lambda row: 
                    all(search_word in ' '.join(map(str, row[text_column1])).lower() 
                        for search_word in search_words) and 
                    all(search_word in ' '.join(map(str, row[text_column2])).lower() 
                        for search_word in search_words), axis=1)

    filtered_df = df[mask]
    sum_int_column1 = filtered_df[int_column1].sum()
    sum_int_column2 = filtered_df[int_column2].sum()
    return len(filtered_df), sum_int_column1, sum_int_column2

def extract_text_between(input_string, start_string, end_string):
    pattern = re.escape(start_string) + r'(.*?)' + re.escape(end_string)
    match = re.search(pattern, input_string, re.DOTALL)
    return match.group(1) if match else ''
def process_substring(input_string):
    no_commas = input_string.replace(',', '')
    words = no_commas.split()
    filtered_words = [word for word in words if 'xa0' not in word]
    clean_string = ' '.join(filtered_words)
    numbers = [int(num) for num in re.findall(r'\b\d+\b', clean_string)]
    return numbers[-1]
def spend_analyzer(profile_index,stri):
    details=det.iloc[profile_index]
    state=extract_state(details['Location'])
    print("Name                    :",details["Name"])
    print("Query                   :",stri)
    print("State                   :",state)
    x=states_to_phase[state]
    if isinstance(x, list) and len(x) > 0:
        x = x[-1]
    start_date=phase_to_nomination_date[x]
    print("Start_Date              :",start_date)
    print("End_Date                :",result_date)
    df_between_dates = filter_ads_by_date_range(df,start_date,result_date)
    x,y,z=process_data(df_between_dates,"twit_ner","spacy_ner","lower_bound_spend","upper_bound_spend",stri)
    print("Number of Satisfying Ads:",x)
    print("Lowest Amount           :",y)
    print("Highest Amount          :",z)
    strx=details["Expenditure"]
    print(strx)
    stry=extract_text_between(strx,"Campaign through electronic/print media","Vehicles used and POL expenditure on such")
    fin=process_substring(stry)
    print("Declared Media Total    :",fin)
```
